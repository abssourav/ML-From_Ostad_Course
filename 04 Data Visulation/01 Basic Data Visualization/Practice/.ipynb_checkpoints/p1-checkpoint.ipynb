```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [10, 20, 30], label="Sales")
plt.bar(['A', 'B', 'C'], [100, 200, 300]) # বার চার্ট লাইন চার্টকে ঢেকে দেবে!
plt.show()
```

```python
fig, ax = plt.subplots(1, 2) # ১টি রো, ২টি কলাম (দুটি আলাদা ঘর)

# প্রথম ঘরে সেলস গ্রাফ
ax[0].plot([1, 2, 3], [10, 20, 30])
ax[0].set_title("Sales Trend")

# দ্বিতীয় ঘরে ক্যাটাগরি গ্রাফ
ax[1].bar(['A', 'B', 'C'], [100, 200, 300])
ax[1].set_title("Category Wise")

plt.tight_layout()
plt.show()
```

```python
fig,axes = plt.subplots(figsize=(10,8))

axes.plot(x,x+2,color='red',linewidth=0.5,ms=15,marker='p')
axes.plot(x,x+1,color='g',lw=1,linestyle=':',marker='p',markersize=10)
axes.plot(x,x+3,color='b',lw=1,linestyle='--',marker='p',markersize=15,markerfacecolor='yellow')
axes.plot(x,x+4,color='g',lw=1,linestyle='-.',marker='h',markersize=12,markerfacecolor='red')
```

```python
import pandas as pd
import seaborn as sns
```

```python
districts = ['Dhaka','Rangpur','Rajsahi','Barishal','Noakhali','Sylhet','Chittagong']
complaint_types = ['Power cute','Traffic','Noise','Traffic','Power Cut','Ice Cream Cart','Garbage','Water Leak']

n_rows = 500

data = {
    'district': np.random.choice(districts,size=n_rows),
    'complaint_types':np.random.choice(complaint_types,size=n_rows),
    'temperatur':np.random.normal(loc=32,scale=5,size=n_rows),
    'electricity_usage':np.random.normal(loc=250,scale=50,size=n_rows),
    'num_ice_cream_carts':np.random.choice([0,1,2,4,10,10,10],size=n_rows),
    'survey_rating':np.random.choice([1,2,3,4,5,np.nan],size=n_rows)
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df.loc[5:10,'complaint_types'] = np.nan
df = pd.concat([df,df.iloc[0:5]])
```

```python
sns.set_theme(style='whitegrid')

plt.figure(figsize=(8,4))
sns.histplot(df['temperatur'],kde=True,bins=10)
plt.title('Temperature distribution')
plt.xlabel('temperature degree')
plt.ylabel('frequency')

plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x='district',y='electricity_usage',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='plasma')

plt.show()
```
